# Welcome to the GeoLibre notebook

This notebook runs entirely in your browser on a [Pyodide](https://pyodide.org) kernel — no server required. Run a cell with **Shift+Enter**.

The `geolibre` module is preloaded, so you can script the **live map next to this notebook** straight from a cell.

```python
import geolibre

m = geolibre.connect()  # a handle to the live map in the app
m
```

## Move the camera

`fly_to` animates to a `(lng, lat)` with an optional zoom.

```python
m.fly_to(-122.4194, 37.7749, zoom=11)  # San Francisco
```

## Add point markers

Pass `(lng, lat)` pairs, or dicts with extra keys that become feature properties.

```python
cities = [
    {"lng": -122.4194, "lat": 37.7749, "name": "San Francisco"},
    {"lng": -74.0060, "lat": 40.7128, "name": "New York"},
    {"lng": -0.1276, "lat": 51.5074, "name": "London"},
    {"lng": 139.6917, "lat": 35.6895, "name": "Tokyo"},
]
m.add_markers(cities, name="World cities")
```

## Add a GeoJSON layer

`add_geojson` accepts a FeatureCollection / Feature / geometry dict, a JSON string, or any object with `__geo_interface__` (e.g. a GeoDataFrame). Install packages on demand with `%pip install geopandas`.

```python
bay_area = {
    "type": "Polygon",
    "coordinates": [
        [[-122.6, 37.6], [-122.6, 37.9], [-122.2, 37.9], [-122.2, 37.6], [-122.6, 37.6]]
    ],
}
m.add_geojson(bay_area, name="Bay Area")
```

## Fit the view to an extent

`fit_bounds` takes `[west, south, east, north]`.

```python
m.fit_bounds([-122.6, 37.6, -122.2, 37.9])
```
